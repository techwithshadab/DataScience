### Multiple Linear Regression-
y = b0 + b1*x1 + b2*x2 + ... + bn*xn <br>
y - dependent variable <br>
x1,x2,...,xn - independent variables <br>
### Assumptions of Linear Regression-
1. Linearity
    
2. Homoscedasticity
3. Multivariate Normality
4. Independence of errors
5. Lack of multicollinearity
<br> These assumptions should be true for a LR model

### 5 Methods of building models:
    1. All-in
        Taking all variables to build the model
        Need to have prior knowldege
        While preparing for backward elimination
        
    2. Backward Elimination      #it is the fastest way among all these
        Selecting significance level to stay in the model (0.05)
        Fitting full model with all possible predictors
        Considering the predictor with highest p-value. If P > SL, go to next step, otherwise go to fin
        Remove the predictor
        Fit the model without this variable # Again go back to step 3 until the p-value becomes smaller then SL
        
    3. Forward Selection         
        Selecting significance level to enter in the model (0.05)
        Fitting all simple linear regression models y ~ xn. Select the one with the lowest P-value
        Keeping that variable and fitting all possible models with one extra predictor added to the one(s) we already have
        Consider the predictor with the lowest P-value. If P<SL, go to 3rd step, else finish and take the previous model
        
    4. Bidirectional Elimination #stepwise regression
        Selecting a significance level to enter and to stay in the model (0.05)
        Performing the next step of forward selection (new variables must have P < SLenter to enter)
        Performing all steps of backward elimination (old variables must have P < SLstay to stay) #go back to step 2
        No new variables can enter or old variables can exit # final model is ready at this step
        
    5. Score Comparison (All possible models)
        Select a criterion of goodness of fit
        Constructing all possible regression models: 2**n -1 total combination
        Selecting the one with the best criterion # 10 columns - 1023 models

### Backward Elimination with p-values only:

```python
import statsmodels.formula.api as sm
def backwardElimination(x, sl):
    numVars = len(x[0])
    for i in range(0, numVars):
        regressor_OLS = sm.OLS(y, x).fit()
        maxVar = max(regressor_OLS.pvalues).astype(float)
        if maxVar > sl:
            for j in range(0, numVars - i):
                if (regressor_OLS.pvalues[j].astype(float) == maxVar):
                    x = np.delete(x, j, 1)
    print(regressor_OLS.summary())
    return x
 
SL = 0.05
X_opt = X[:, [0, 1, 2, 3, 4, 5]]
X_Modeled = backwardElimination(X_opt, SL)
```
